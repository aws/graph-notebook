## PoC Problem to solve (use case) based on open source data 

In the network domain, Operators need to know to which cell a user is or will be connected. i.e Operators need to know how many users a given cell is serving to adjust capacity, to detect congestion, to anticipate degradation etc. 

**How it works: User and Cell attachement (simplified)**
- A cell in a cellular network is a geographic area that is covered by a base station or cell tower. 
- The size of a cell can vary depending on the density of the population and the terrain of the area. 
- Mobile devices communicate with the base station in their respective cell to make calls, send messages, and access the internet. 
- As a user moves from one cell to another, the call or data session is handed off to the next cell seamlessly, ensuring uninterrupted connectivity.

## What is a Link prediction

1. Link Prediction is a task in graph analysis where the goal is to predict missing or future connections between nodes based on their attributes.
2. One of the key task in Graph and Key driver for graph approach adoption
3. Fully non supervised
4. takes nodes properties as well as edge properties to **guess** the relationships between nodes.
5. Unlimitted applications

## Link prediction :  a machine learning task 
**Link prediction algorithms** can be used to predict the likelihood of a link between a gNodeB and its cells to be serving a UE. 
As stated above, this information can be used to improve the performance of 5G networks: 

- **Optimize the placement of gNodeBs**: network operators can optimize the placement of gNodeBs to improve the performance of the network.
- **Predict the performance of the network**: network operators can predict the performance of the network under different load conditions.
- **Identify potential problems**: network operators can identify potential problems, such as overloaded gNodeBs or low-quality links.
- **Energy efficiency**: track the unused or low used cells / gnodeB

## PoC- Dataset description 

**Open source dataset referenced by Stanford ** 

- https://snap.stanford.edu/data/#tcn

- referenced by **Stanford**. It is provided by Global Technology Services department of Huawei Technologies, Co., Ltd. It represents **the relationship between users and behaviors in Telecom network** and is used for the evaluation of standard of modern graph algorithms on Telecom networks. 

- **anonimized and normalized** features


**Description of the dataset**
- The telecom graph consists of 4 types of nodes (**user, package, app, cell**) and 3 types of edges (**user-buy-package, user-use-app, user-live-cell**).

- There are **170K nodes** and **50M edges**, among which **100K are users**, and all files are plain texts in CSV format, with the file name beginning with 'edge_' or 'node_'.

- For **node files**, each row is:nodeID,attr1,attr2,...
- For **edge files**, each row is:srcID,dstID,attr1,attr2,...
- All **IDs** are 4-byte integers, while attributes are all 4-byte float numbers within [0.f,1.f].

For the PoC about Cell prediction, **Cell** and **User** nodes are used as well as the **Edge**: **user_live_cell**



**In this blog the data we are using is opensource data where all attributes are anonimized and without temporal attributes**
